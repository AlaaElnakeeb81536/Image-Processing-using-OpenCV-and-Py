# Thresholding

```python
#obtain the optimal threshold value
thresh=100

#apply thresholding to image
binary = grayscale > thresh

#sgow image
show_image(grayscale,'original')
show_image(binary,'thresholded')
```

```python
#obtain the optimal threshold value
thresh=100

#apply thresholding to image
inverted_binary = grayscale <= thresh

#sgow image
show_image(grayscale,'original')
show_image(inverted_binary,'thresholded')
```

# Another way for thresholding

```python
from skimage.filters import try_all_threshold

fig,ax=try_all_threshold(grayscale,verbose=False)

plt.show(fig,ax)
```

# OPtimal thresh value
Global==>Uniform background

```python
#import the ostu threshold function

from skimage.filters import threshold_otsu

#obtain the optimal threshold value

thresh=threshold_otsu(grayscale)

#apply thresholding to the image
binary_global=grayscale>thresh
```

```python
show_image(grayscale,'original')
show_image(binary_global,'thresholed')
```

# Local ==> uneven background

```python
from skimage.filters import threshold_local

#set the blocksize to 35
block_size=35

#obtain the optimal threshold value
local_thresh=threshold_local(grayscale,block_size,offset=10)

#apply thresholding to the image
binary_local=grayscale>local_thresh
```

```python
show_image(grayscale,'original')
show_image(binary_local,'thresholded')
```

# Sobel
before applying this for edge detection ,image should be in grayscale

```python
def plot_comparison(original,filtered,title_filtered):

    fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(8,6),sharex=True
                               ,sharey=True)
    ax1.imshow(original,cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered,cmap=plt.cm.gray)
    ax2.set_title(title_filtered)
    ax2.axis('off')
```

```python
#import module and function
from skimage.filters import sobel

#apply edge detection filter
edge_soble=sobel(original)

plot_comparison(original,edge_soble,'filtered_image')
```

# 2-Gaussian smoothing

used to blur an image or to reduce noise

```python
image=plt.imread(r'C:\Users\magic\Pictures\smoothing.png')
```

```python
#import the module and function
from skimage.filters import gaussian

#apply edge detection filter
gaussian_image=gaussian(image,multichannel=False) #false if image in grayscale ,True if image in color

#show original and resulting image to compare
plot_comparison(image,gaussian_image,'blurred with gaussian filter')
```
